Step 5 : Now  we have prepared the data for our model

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,1].values
# print(x)
```

#Splitting the data
Step 6 : Now, we have divided the data for training & testing the model 


```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

Model

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
print("Training Completed")
```

Step 8:Now our model is ready . test prediction is correct or not 

```python
print(y_test)
print("Prediction of scores")
y_pred=model.predict(x_test)
print(y_pred)


```

step 8:Now check the acurracy of model

```python
df=pd.DataFrame({'Actual':y_test,'Predicited':y_pred})
```

```python
df
```

Best line that fits our data


```python
sns.regplot(data=df,x=x,y=y)
```

What will be the predicted score when a student studies 9.25 hrs/ day?

```python
hours=[[9.25]]
pred=model.predict(hours)
print(pred)
```

Evaluating the model

```python
from sklearn import metrics
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
```
